```python
# importing libraries 
import pandas as pd # for strings data
import numpy as np # for numerical operations
import matplotlib.pyplot as plt # for plots and visulations
```

```python
# loading and reading the dataset
df=pd.read_excel("Data Analyst Assignment (1).xlsx",sheet_name='Assignment-2')
```

# Monthly and Quarterly sales data

```python
# selecting only category columns ie.(SKU) from the data
b1=df.iloc[:,:1]
```

```python
# selecting the date columns from the dataset
bn=df.iloc[:,1:]
```

```python
# converting date columns into quaterly format
a11=bn.groupby(pd.PeriodIndex(bn.columns,freq='Q'),axis=1).sum()
```

```python
#creating quaterly file from quater columns and SKU column
quarterly=pd.concat([b1,a11],axis=1)
```

```python
# quaterly sales data
quarterly
```

```python
# converting date columns into monthly format
a12=bn.groupby(pd.PeriodIndex(bn.columns,freq='M'),axis=1).sum()
```

```python
# creating quaterly file from month columns and SKU column
monthly=pd.concat([b1,a12],axis=1)
```

```python
monthly
```

# Top 3 Months in terms of overall sales

```python
# top 3 months by total sales
monthly.sum().sort_values(ascending=False).head(3)
```

# Outliers in sales

```python
# importing statistic library
from scipy import stats
z=np.abs(stats.zscore(monthly.iloc[:,1:2]))  # selecting auguest month of 2019
```

```python
# selecting ouliers present in auguest 2019 data
ot=pd.DataFrame(z>3)
```

```python
# creating oulier file by combining outlier data and auguest 2019 data
outliers=pd.concat([monthly.iloc[:,0:2],ot],axis=1)
```

```python
# selecting the outliers 
outliers.loc[outliers[0]==True]
```
